#Histograms and Polar Plots

In this unit, we would like you to create **histograms** and **polar plots** to display the data.

```python
#Helper code we need to get started
%matplotlib inline
import numpy as np
import matplotlib.pyplot as p
```

```python
data = np.random.rand(1000)
```

```python
p.plot(data)
```

```python
p.hist(data)
```

```python
datan= np.random.randn(1000)
p.hist(datan)
```

```python
help(p.hist)
```

```python
p.hist(datan, 20,cumulative=True)
```

```python
p.hist(datan, 20,normed=True,cumulative=True,color='r')
p.xlabel('Random Normalized Data')
p.ylabel('Number of points')
p.title('Cumulative Density Function of the Data')
```

```python
x = np.arange(0, 360, 45)
x
```

```python
spikecounts = np.random.rand(8)*20
```

```python
p.bar(x,spikecounts)
```

```python
help(p.bar)
```

```python
p.bar(x,spikecounts,width=45)
```

```python
p.bar(x,spikecounts,width=45)
p.xlabel('Direction')
p.ylabel('Spike Counts')
p.title('Tuning Curve')
p.axis([0, 360, 0, 14])
```

```python
p.polar(spikecounts)
```

```python
help(p.polar)
```

```python
p.polar(x*np.pi/180,spikecounts)
```

```python
spikecounts2 = np.append(spikecounts,spikecounts[0])
r = np.arange(0, 361, 45)*np.pi/180
p.polar(r,spikecounts2)
p.title('Tuning Curve')
```
